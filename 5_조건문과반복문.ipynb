# 4. continue문과 break문

```python
#continue는 계속해서 반복문 진행
#break는 중지하고 반복문을 끝냄
absent = [2,5] #결석
no_book = [7] #책을 안가져옴
for student in range(1,11):
    if student in absent:
        continue
    elif student in no_book:
        print("오늘 수업 여기까지. {}는 교무실로 따라와".format(student))
        break
    print("{}, 책을 읽어봐".format(student))
```

# 5. 한줄for문
